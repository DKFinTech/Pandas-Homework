### Determine which portfolios are riskier than the S&P 500

```python
# Calculate  the daily standard deviation of S&P 500
Cumulative_Returns['S&P 500'].std()
# Determine which portfolios are riskier than the S&P 500
for x in Cumulative_Returns:
    if Cumulative_Returns[x].std()  > Cumulative_Returns['S&P 500'].std():
        print(f"{x} is riskier than S&P 500")
```

### Calculate the Annualized Standard Deviation

```python
# Calculate the annualized standard deviation (252 trading days)
Variation = Cumulative_Returns.var()
Mean = Variation.mean() + 1
Annualized_std_dev = np.sqrt(((Variation + (Mean**2))**12) - Mean**24)
Annualized_std_dev
```

## Rolling Statistics Challenge: Exponentially Weighted Average 

An alternative way to calculate a rolling window is to take the exponentially weighted moving average. This is like a moving window average, but it assigns greater importance to more recent observations. Try calculating the [`ewm`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.ewm.html) with a 21-day half-life.

```python
# Use `ewm` to calculate the rolling window
Exp_Weighted_Avg = Cumulative_Returns.ewm(halflife = 21)
Exp_Weighted_Avg.mean()
```

# Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. After all, if you could invest in one of two portfolios, and each offered the same 10% return, yet one offered lower risk, you'd take that one, right?

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Annualized Sharpe Ratios
Sharpe_Ratio = (Annualized_std_dev.mean() * 252) / (Annualized_std_dev * np.sqrt(252))
Sharpe_Ratio.head()
```

```python
# Visualize the sharpe ratios as a bar plot
Sharpe_Ratio.plot.bar(title='Sharpe Ratios')
```

```python
# Reorganize portfolio data by having a column per symbol
```

```python

# Calculate daily returns & drop NAs
Second_Portfolio_Dates = pd.to_datetime(Second_Portfolio_Reset_Index['Date'])
AMD_Daily_Returns = Second_Portfolio_Reset_Index['AMD'].pct_change().dropna()
TWTR_Daily_Returns = Second_Portfolio_Reset_Index['Twitter'].pct_change().dropna()
F_Daily_Returns = Second_Portfolio_Reset_Index['Ford Motor'].pct_change().dropna()

Second_Portfolio_Daily_Returns = pd.concat([Second_Portfolio_Dates, AMD_Daily_Returns,TWTR_Daily_Returns, F_Daily_Returns], axis = 1).dropna().reset_index().drop(columns = 'index')

# Display sample data
Second_Portfolio_Daily_Returns.head()
```

## Calculate the weighted returns for the portfolio assuming an equal number of shares for each stock

```python
# Set weights
weights = [1/3, 1/3, 1/3]

#Remove date column to prevent shape error
Weighted_Portfolio = Second_Portfolio_Daily_Returns.drop(columns = 'Date')

# Calculate portfolio return
Weighted_Returns = Weighted_Portfolio.dot(weights)

#Add back the date column
Numerical_Date_Column = pd.to_datetime(Second_Portfolio_Daily_Returns['Date'])
Weighted_Returns_With_Date_Column = pd.concat([Numerical_Date_Column, Weighted_Returns], axis = 1).dropna().reset_index()

#Remove the extra index column by 'spliting' it from the datafram
Split_Weighted_Returns = Weighted_Returns_With_Date_Column.drop(columns = 'index')

# Display sample data
Split_Weighted_Returns = Split_Weighted_Returns.rename(columns = {0: 'Weighted'})
Split_Weighted_Returns.head()
```
